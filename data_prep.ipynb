```python
import glob
import random
import os
# -------- REAL (WAV) --------
if os.name == "nt":  # Windows
    cv_path = r"D:\Alpha\ML\subset_data\wav_data\real"
else:  # Linux
    cv_path = "/mnt/D/Alpha/ML/subset_data/wav_data/real"

cv_files = glob.glob(os.path.join(cv_path, "*.wav"))

print("Total REAL wav files:", len(cv_files))


# -------- SPOOF (WAV) --------
if os.name == "nt":  # Windows
    spoof_path = r"D:\Alpha\ML\subset_data\wav_data\spoof"
else:  # Linux
    spoof_path = "/mnt/D/Alpha/ML/subset_data/wav_data/spoof"

spoof_files = glob.glob(os.path.join(spoof_path, "*.wav"))

print("Total SPOOF wav files:", len(spoof_files))

# -------- BALANCE PROPERLY --------
n = min(len(cv_files), len(spoof_files))

random.shuffle(cv_files)
random.shuffle(spoof_files)

real_sample = cv_files[:n]
spoof_sample = spoof_files[:n]
# -------- COMBINE --------
file_list = real_sample + spoof_sample
labels = [0]*n + [1]*n

# -------- SHUFFLE --------
combined = list(zip(file_list, labels))
random.shuffle(combined)

file_list, labels = zip(*combined)
file_list = list(file_list)
labels = list(labels)

# -------- FINAL CHECK --------
print("Final dataset size:", len(file_list))
```
